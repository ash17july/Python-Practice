```python
from tkinter import *
root = Tk()
root.geometry('400x300')
root.minsize(300,100)
root.maxsize(1200,988)
f=Frame(root,bg='pink',borderwidth=6, relief=RIDGE)
f.pack(side=LEFT)
f1=Frame(root,bg='pink',borderwidth=6, relief=RIDGE)
f1.pack(side=RIGHT)
f2=Frame(root,bg='pink',borderwidth=6, relief=RIDGE)
f2.pack(side=BOTTOM)
f3=Frame(root,bg='pink',borderwidth=6, relief=RIDGE)
f3.pack(side=TOP)
f4=Frame(root,bg='pink',borderwidth=6, relief=RIDGE)
f4.pack()
f5=Frame(root,bg='pink',borderwidth=6, relief=RIDGE)
f5.pack(side=TOP, anchor='ne')
f6=Frame(root,bg='pink',borderwidth=6, relief=RIDGE)
f6.pack(side=TOP, anchor='nw')
f7=Frame(root,bg='pink',borderwidth=6, relief=RIDGE)
f7.pack(side=LEFT, anchor='sw')
f8=Frame(root,bg='pink',borderwidth=6, relief=RIDGE)
f8.pack(side=RIGHT, anchor='se')
def red():
    print("it is a red color")
def yellow():
    print("It is a yellow color")
def orange():
    print("It is an orange color")
def green():
    print("It is a green color")
def pink():
    print("It is a pink color")
    
def brown():
    print("It is a brown color")
def purple():
    print("It is an purple color")
def violet():
    print("It is a violet color")
def grey():
    print("It is a grey color")
    
b = Button(f3, fg='red', text='red', command=red)
b.pack(side=TOP)
b1 = Button(f, fg='yellow', text='yellow', command=yellow)
b1.pack(side=LEFT)
b2 = Button(f1, fg='green', text='green', command=green)
b2.pack(side=RIGHT)
b3 = Button(f2, fg='orange', text='orange', command=orange)
b3.pack(side=BOTTOM)
b4 = Button(f4, fg='pink', text='pink', command=pink)
b4.pack()
b5 = Button(f5, fg='brown', text='brown', command=brown)
b5.pack(side=TOP, anchor = 'ne')
b6 = Button(f6, fg='purple', text='purple', command=purple)
b6.pack(side=LEFT, anchor = 'nw')
b7 = Button(f7, fg='violet', text='violet', command=violet)
b7.pack(side=BOTTOM, anchor= 'sw')
b8 = Button(f8, fg='grey', text='grey', command=grey)
b8.pack(side=BOTTOM, anchor='se')
root.mainloop()
```
